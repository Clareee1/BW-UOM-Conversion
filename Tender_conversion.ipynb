```python
import pandas as pd
import numpy as np
###---name might change on customer UoM---###
cus_uom,bw_uom = df1["UoM"],df1["Price UOM"]

##iterate the column name
#for col in df1.columns:
 #   print(col)
    
###---"Packing""UoM""Price UOM" name might be different---###
df2 = df1[["Sort Key","Item description","Part Description","Packing","UoM","Price UOM","Part Number"]]
##Assume all conversion is 1 currently
df2.loc[:, "conversion"] = 1
###helping function: store the information from the description
df2.loc[:, "Packing_qty"] = 0

##Each units for customer UoM
#when customer UoM = each with BW UoM = each 
#Packing description is not null
each = df2.loc[df2["UoM"].isin(["EA","Each","ea","EACH"]) & df2["Price UOM"].isin(["EA","Each","ea","EACH"])& df2["Packing"].notnull()]
#when packing description is not null
each.head()
```

```python
#pk_qty_convert: packing quantity convert
pk_qty = each.assign(Packing_qty = lambda x: x['Packing'].str.findall(r'(\d+)'))
pk = pk_qty[pk_qty['Packing_qty'].str.len() == 1]
pk['Packing_qty'] = pk['Packing_qty'].str.get(0)
pk['conversion'] = pk['Packing_qty'].fillna(pk['Packing_qty'])
pk.head()
```

```python
##replace the conversion
df2.update(pk)
rslt_df = df2[(df2['Sort Key'] >= 40) & (df2['Sort Key'] <= 45)]
##transfer example
rslt_df
```

```python
##Each_unmatch units for customer UoM
#when customer UoM = each with BW UoM != each 
each_unmatch = df2.loc[df2["UoM"].isin(["EA","Each","ea","EACH"]) & (~df2["Price UOM"].isin(["EA","Each","ea","EACH"]))]

#filling in packing_qty if is not null
each_unmatch = each_unmatch.assign(Packing_qty=np.where(each_unmatch.Packing.isnull(), each_unmatch.Packing_qty, (lambda x: x['Packing'].str.findall(r'(\[0-9]+)'))))
for index, row in each_unmatch.iterrows():
    if row['Packing_qty'] != 0 and row['Packing_qty'].str.len() == 1:
        row['Packing_qty'] = row['Packing_qty'].str.get(0) 

each_unmatch.head(5)
```
